```python
import pandas as pd

# Load the dataset (adjust path as needed)
df = pd.read_csv("C:\Customer Segmentation Analysis - Copy.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: group by age and sum revenue
age_revenue = df.groupby('age')['price'].sum().reset_index()

plt.figure(figsize = (12, 6))

sns.lineplot(data = age_revenue, x = 'age', y = 'price', marker = 'o', color = 'purple')

plt.title('Total Revenue by Age Group')

plt.xlabel('Age')

plt.ylabel('Total Revenue')

plt.grid(True)

plt.show()
```

```python
# Count number of categories per age
category_count = df.groupby('age')['category'].count().reset_index(name = 'Category Count')

plt.figure(figsize = (12, 6))

sns.lineplot(data = category_count, x = 'age', y = 'Category Count', color = 'blue', marker = 'o')

plt.title('Sum of Age by Category')

plt.xlabel('Age')

plt.ylabel('Count of Category')

plt.grid(True)

plt.show()
```

```python
payment_counts = df['payment_method'].value_counts()

plt.figure(figsize = (8, 8))

plt.pie(payment_counts, labels = payment_counts.index, autopct = '%1.1f%%', startangle = 140, colors = sns.color_palette("pastel"))

plt.title('Payment Method Distribution')

plt.show()
```

```python
gender_revenue = df.groupby('gender')['price'].sum()

plt.figure(figsize = (8, 8))

plt.pie(gender_revenue, labels = gender_revenue.index, autopct = '%1.2f%%', startangle = 140, colors = ['orchid', 'skyblue'])

plt.title('Total Revenue by Gender')

plt.show()
```
